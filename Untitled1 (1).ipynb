```python
df = pd.read_csv('sample_df.csv')
df.head()
```

```python
import sqlite3
conn = sqlite3.connect(":memory:")
df.to_sql("ecommerce", conn, index=False, if_exists='replace')

```

```python
pd.read_sql_query("SELECT * FROM ecommerce LIMIT 5;", conn)
```

```python
pd.read_sql_query("SELECT event_type, COUNT(*) AS total_events FROM ecommerce GROUP BY event_type;", conn)
```

```python
pd.read_sql_query("SELECT round(AVG(price),2) AS avg_price FROM ecommerce;",conn)
```

```python
pd.read_sql_query("SELECT brand,ROUND(SUM(price),2) AS total_revenue FROM ecommerce WHERE brand IS NOT NULL GROUP BY brand ORDER BY total_revenue DESC LIMIT 5;",conn)
```

```python
pd.read_sql_query("SELECT user_id, COUNT(*) AS total_actions FROM ecommerce GROUP BY user_id ORDER BY total_actions DESC LIMIT 5;",conn)
```

```python
pd.read_sql_query("SELECT * FROM ecommerce WHERE price > (SELECT AVG(price) FROM ecommerce);",conn)
```

```python
pd.read_sql_query("SELECT category_code, MAX(price) AS max_price FROM ecommerce WHERE category_code IS NOT NULL GROUP BY category_code;",conn)
```

```python
pd.read_sql_query("SELECT COUNT(DISTINCT brand) AS total_brands FROM ecommerce;",conn)
```

```python
pd.read_sql_query("SELECT a.user_id, a.event_type AS first_event, b.event_type AS next_event FROM ecommerce a JOIN ecommerce b ON a.user_id = b.user_id AND a.user_session = b.user_session AND a.event_time < b.event_time LIMIT 5;", conn)
```
